```python
primes = []

for num in range(2, 21): 
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):  
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

alternate_sum = sum(primes[::2]) 

print("Sum of alternate prime numbers from 1 to 20:", alternate_sum)
```

```python
def move_zeros(nums):
    pos = 0  # Marks position for next non-zero element
    for num in nums:
        if num!= 0:
            nums[pos] = num
            pos += 1
    # Fill the rest with zeros
    for i in range(pos, len(nums)):
        nums[i] = 0
```
